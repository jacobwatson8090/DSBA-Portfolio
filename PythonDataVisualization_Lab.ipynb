```python
import seaborn as sns
import pandas as pd
df = pd.read_csv("Amazon.csv")
```

```python
new = df["category"].str.split("|", expand = True)
print(new)
```

```python
df["Category - 1"]= new[0]
df["Category - 2"]= new[1]
df["Category - 3"]= new[2]
df["Category - 4"]= new[3]
df["Category - 5"]= new[4]
```

```python
df.drop(columns =["category"], inplace = True)
```

```python
df
```

```python
categories_to_change = ['Electrical', 'Home Medical Supplies & Equipment', 'Microphones', 'Car Accessories', 'Office Paper Products', 'Office Electronics', 'Arts & Crafts']


df.loc[df['Category - 2'].isin(categories_to_change), 'Category - 1'] = 'All Other'

df
```

```python
median_discount_per_category2 = df.groupby('Category - 2')['discount_percentage'].median().reset_index()

print(median_discount_per_category2)
```
